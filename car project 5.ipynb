```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
df
```

```python
df.info()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='seller_type',data=df,palette='viridis')
plt.title('selling_type count plot')
plt.xlabel('seller_type')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True,color='yellow')
plt.title('Hisplot for year')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
cars=df['owner'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(cars,labels=['First Owner','Second Owner','Third Owner','Fourth & Above Owner','Test Drive Car'],
        autopct='%1.1f%%',colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#ff99cc'],
        startangle=180,explode=(0.05,0.05,0.05,0.05,0.05))


plt.title('Owner pie Chart')
plt.show()
```

```python
cars=df['owner'].value_counts()
cars
```

```python
#boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='seller_type',y='selling_price',data=df,palette='Set2')
plt.title('Fare Distribution by passenger class')
plt.xlabel('class')
plt.ylabel('Fare')
plt.show()
 
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='year',y='km_driven',hue='owner',
                style='seller_type',data=df,palette='rainbow')
plt.title('year vs km_driven with owner and seller_type')
plt.xlabel('year')
plt.ylabel('km_driven')
plt.legend(title='about Car')
plt.show()
```

```python
df.info()
```

```python
num=df[['year','selling_price','km_driven']].corr()
num
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(num,annot=True,cmap='rainbow',center=0)
plt.title('Correlation Hearmap of Car Dataset')
plt.show()
```

```python
#violin plot
plt.figure(figsize=(10,8))
sns.violinplot(x='fuel',y='year',hue='seller_type',data=df,split=True,palette='Set2')
plt.title('seller_type Distribution by year and fuel')
plt.xlabel('fuel')
plt.ylabel('year')
plt.legend(title='Survived')
plt.show()

```

```python
##KDE
sns.kdeplot(data=df['year'].dropna(),color='red',shade=True)
plt.title('year Distribution (Kernel Density Estimate)')
plt.xlabel('year')
plt.ylabel('Density')
plt.show()
```

```python
#frace grid plot
g=sns.FacetGrid(df,col='seller_type',row='owner',margin_titles=True)
g.map(sns.histplot,'year',bins=20,color='teal')
g.fig.suptitle('year distribution by seller type and owner',y=1.05)
plt.show()
```

```python
#pairplot
sns.pairplot(df[['year','selling_price','km_driven','transmission']],
             hue='transmission',palette='husl',
             plot_kws={'alpha':0.6},diag_kind='kids')
plt.suptitle('Pair plot of titanc variables',y=1.02)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.swarmplot(x='fuel',y='km_driven',hue='transmission',
              data=df,palette='Set1',dodge=True)
plt.title('transmission distribution by fuel and km_driven(swarm plot)')
plt.xlabel('fuel')
plt.ylabel('km_driven')
plt.legend(title='transmission')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.swarmplot(x='fuel',y='km_driven',hue='transmission',
              data=df,palette='Set1',dodge=True)
plt.title('transmission distribution by fuel and km_driven(swarm plot)')
plt.xlabel('fuel')
plt.ylabel('km_driven')
plt.legend(title='transmission')
plt.show()
```
